```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is already defined and loaded with data
# Uncomment the next line if you need to load data from a CSV
# df = pd.read_csv('path_to_your_file.csv')

# Inspect the DataFrame
print("DataFrame Head:")
print(df.head())
print("\nDataFrame Columns:")
print(df.columns)

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Data cleaning steps
def clean_numeric_columns(df, columns):
    for column in columns:
        df[column] = df[column].astype(str).str.replace(',', '')
        df[column] = pd.to_numeric(df[column], errors='coerce')
    # Replace NaN values with 0 for numerical columns
    df.fillna(0, inplace=True)
    return df

# Define columns to clean
numeric_columns = ['Revenue (USD millions)', 'Employees']
# Clean numeric columns and handle NaNs
df = clean_numeric_columns(df, numeric_columns)

# Re-check summary statistics after cleaning
print("\nSummary Statistics after Cleaning:")
print(df.describe())

# Ensure the 'Industry' column exists before proceeding
if 'Industry' in df.columns:
    # Group by Industry and calculate average revenue and total employees
    industry_group = df.groupby('Industry').agg({
        'Revenue (USD millions)': 'mean',
        'Employees': 'sum'
    }).reset_index()
    print("\nGrouped by Industry:")
    print(industry_group)
else:
    print("\nThe column 'Industry' does not exist in the DataFrame.")

# Find the company with the highest revenue
if 'Revenue (USD millions)' in df.columns:
    max_revenue_company = df[df['Revenue (USD millions)'] == df['Revenue (USD millions)'].max()]
    print("\nCompany with the Highest Revenue:")
    print(max_revenue_company)
else:
    print("\nThe column 'Revenue (USD millions)' does not exist in the DataFrame.")

# Check if 'Name' column exists before plotting
if 'Name' in df.columns and 'Revenue (USD millions)' in df.columns:
    plt.figure(figsize=(20, 16))

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.histplot(df['Revenue (USD millions)'], kde=True, color='skyblue')
plt.title('Revenue Distribution of the Largest US Companies')
plt.xlabel('Revenue (USD millions)')
plt.ylabel('Frequency')
plt.show()
```

```python
industry_group = df.groupby('Industry')['Revenue (USD millions)'].mean().sort_values(ascending=False)

plt.figure(figsize=(14, 7))
industry_group.plot(kind='bar', color='teal')
plt.title('Average Revenue by Industry')
plt.xlabel('Industry')
plt.ylabel('Average Revenue (USD millions)')
plt.show()
```
